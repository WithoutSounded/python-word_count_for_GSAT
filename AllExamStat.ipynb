```python
dataPath = 'D:\\Justin\'s_University\\專題(learning)\\output'
wholeFile = os.listdir(dataPath)
```

```python
AllWordStat = {}
ctr = 0

for file in wholeFile:
    File = dataPath+'\\'+file
    
    with open(File, newline='') as csvfile:
        rows = csv.DictReader(csvfile)
        for row in rows:
            word = row
#--------------------------------------------------------------------------------------------------------------------------
            if word['word'] in AllWordStat:
                AllWordStat[word['word']]['total'] = str(int(AllWordStat[word['word']]['total'])+int(word['total']))
                if ctr == 0:
                    AllWordStat[word['word']]['ACT'] = word['total']
                elif ctr == 2:
                    AllWordStat[word['word']]['GSAT'] = word['total']
                elif ctr == 1:
                    AllWordStat[word['word']]['BCT'] = word['total']
                elif ctr == 3:
                    AllWordStat[word['word']]['TVE'] = word['total']
                elif ctr == 4:
                    AllWordStat[word['word']]['TVE42'] = word['total']
                else:
                    pass

            else:
                if ctr == 0:
                    AllWordStat[word['word']] = {'word':word['word'], 'total':word['total'], 'ACT':word['total'], 'GSAT': '', 'BCT': '', 'TVE': '', 'TVE42': ''}
                elif ctr == 2:
                    AllWordStat[word['word']] = {'word':word['word'], 'total':word['total'], 'ACT':'', 'GSAT': word['total'], 'BCT': '', 'TVE': '', 'TVE42': ''}
                elif ctr == 1:
                    AllWordStat[word['word']] = {'word':word['word'], 'total':word['total'], 'ACT':'', 'GSAT': '', 'BCT': word['total'], 'TVE': '', 'TVE42': ''}
                elif ctr == 3:
                    AllWordStat[word['word']] = {'word':word['word'], 'total':word['total'], 'ACT':'', 'GSAT': '', 'BCT': '', 'TVE': word['total'], 'TVE42': ''}
                elif ctr == 4:
                    AllWordStat[word['word']] = {'word':word['word'], 'total':word['total'], 'ACT':'', 'GSAT': '', 'BCT': '', 'TVE': '', 'TVE42': word['total']}
                else:
                    pass
#--------------------------------------------------------------------------------------------------------------------------
    ctr += 1
```
